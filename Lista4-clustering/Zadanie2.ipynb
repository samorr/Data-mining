```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans, Birch, DBSCAN
```

```python
X1 = np.random.rand(1000,2) * np.array([5,100])
X2 = np.random.rand(1000,2) * np.array([5,100]) + np.array([10, 0])
```

```python
X = np.append(X1,X2,axis=0)
```

```python
def myPlotClustering(X, plt_labels):
    plt_K = np.unique(plt_labels).size
    c = ['b','g','r','c','m','y','k','w']
    for k in range(plt_K):
        mask = (plt_labels == k)
        plt.plot(X[mask, 0], X[mask, 1], 'ko', markerfacecolor=c[k])

    plt.plot(X[plt_labels==-1, 0], X[plt_labels==-1, 1], 'ko')
    plt.show()
```

```python
plt.plot(X[:,0], X[:,1], 'r.')
plt.xlim([-30,50])
plt.show()
```

```python
k_means = KMeans(init='k-means++', n_clusters=2, n_init=1)
k_means.fit(X)
```

```python
myPlotClustering(X, k_means.labels_)
```

```python
normX = (X - np.min(X,axis=0))/(np.max(X,axis=0) - np.min(X,axis=0))
plt.plot(normX[:,0], normX[:,1], 'r.')
plt.show()
```

```python
k_means = KMeans(init='k-means++', n_clusters=2, n_init=1)
k_means.fit(normX)
```

```python
myPlotClustering(normX, k_means.labels_)
```

```python
standX = (X - np.mean(X,axis=0))/np.std(X,axis=0)
plt.plot(standX[:,0], standX[:,1], 'r.')
plt.show()
```

```python
k_means = KMeans(init='k-means++', n_clusters=2, n_init=1)
k_means.fit(standX)
```

```python
myPlotClustering(normX, k_means.labels_)
```

```python
birch = Birch(threshold=10, n_clusters=None)
birch.fit(X)
```

```python
myPlotClustering(X, birch.labels_)
```

```python
birch = Birch(threshold=10, n_clusters=2)
birch.fit(X)
```

```python
myPlotClustering(X, birch.labels_)
```

```python
birch = Birch(threshold=0.25, n_clusters=2)
birch.fit(normX)
```

```python
myPlotClustering(normX, birch.labels_)
```

```python
birch = Birch(threshold=0.25, n_clusters=2)
birch.fit(standX)
```

```python
myPlotClustering(standX, birch.labels_)
```

```python
dbscan = DBSCAN(eps=2, min_samples=5)
dbscan.fit(X)
```

```python
myPlotClustering(X, dbscan.labels_)
```
