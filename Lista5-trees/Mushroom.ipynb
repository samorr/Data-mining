```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn import tree
import numpy as np
import pandas as pd
from sklearn import metrics
```

```python
data = pd.read_csv('/home/dominik/Dokumenty/Studia/Data-mining/Lista5-trees/mushroom.data', names=['class', 'cap-shape', 'cap-surface', 'cap-color', 'bruises?', 'odor', 'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color', 'stalk-shape', 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat'])
data.head()
```

```python
data = data.drop('stalk-root',axis=1)
```

```python
data.head()
```

```python
for name in data.columns.values:
    data[name] = pd.Categorical(data[name]).codes
```

```python
data.head()
```

```python
mask = np.random.uniform(0,1, len(data))
train_data = data[mask <= 0.7]
test_data = data[mask > 0.7]

y = train_data['class']
X = train_data.drop('class', axis=1)
print(X.shape)

t = tree.DecisionTreeClassifier()
t = t.fit(X,y)

y = test_data['class']
X = test_data.drop('class', axis=1)
print(X.shape)


t.score(X,y)
```

```python
importances = t.tree_.compute_feature_importances()
importances
```

```python
clearMask = [i+1 for i in range(len(importances)) if importances[i] < 0.01]
clearedData = data.drop(data.columns.values[clearMask],axis=1)
```

```python
mask = np.random.uniform(0,1, len(clearedData))
train_data = clearedData[mask <= 0.7]
test_data = clearedData[mask > 0.7]

y = train_data['class']
X = train_data.drop('class', axis=1)
print(X.shape)

t = tree.DecisionTreeClassifier()
t = t.fit(X,y)

y = test_data['class']
X = test_data.drop('class', axis=1)
print(X.shape)


t.score(X,y)
```

```python
confusion_matrix = metrics.confusion_matrix(y, t.predict(X))
print ("Confusion Matrix\n", confusion_matrix)
```

```python
data['train'] = np.random.uniform(0, 1, len(data))
err = 0
for i in range(10):
    data_test = data[(data['train'] >= i*0.1) & (data['train'] < (i+1)*0.1)]
    data_train = data[(data['train'] < i*0.1) | (data['train'] >= (i+1)*0.1)]

    y = data_train['class']
    X = data_train.drop('class', axis = 1).drop('train',axis=1)

    t = tree.DecisionTreeClassifier()
    t = t.fit(X, y)
#     print('depth', [i.tree_.max_depth for i in t.estimators_])
    y = data_test['class']
    X = data_test.drop('class', axis = 1).drop('train',axis=1)
    print(t.score(X,y))
    confusion_matrix = metrics.confusion_matrix(y, t.predict(X))
    print ("Confusion Matrix\n", confusion_matrix)
    err += t.score(X,y) / 10

print('Average score: ', err)
```
