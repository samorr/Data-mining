```python
% matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from sklearn import (decomposition, manifold)
```

```python
import scipy.io
X = scipy.io.loadmat('Lista07Dane/ReducedImagesForTraining.mat')['images'].T
X.shape
labels = (np.arange(1,251) - 1 ) // 5 + 1
```

```python
X_pca = decomposition.PCA(n_components=2).fit_transform(X)
plt.scatter(X_pca[:,0], X_pca[:,1],c=labels)
```

```python
X_kpca = decomposition.KernelPCA(n_components=2, kernel="poly", degree=2, gamma=10).fit_transform(X)
plt.scatter(X_kpca[:,0], X_kpca[:,1],c=labels)
```

```python
X_iso = manifold.Isomap(30, n_components=2).fit_transform(X)
plt.scatter(X_iso[:,0], X_iso[:,1],c=labels)
```

```python
X_lle = manifold.LocallyLinearEmbedding(30, n_components=2, reg=0.0001).fit_transform(X)
plt.scatter(X_lle[:,0], X_lle[:,1],c=labels)
```

```python
X_mds = manifold.MDS(n_components=2, eps=1e-5,n_init=1, max_iter=100).fit_transform(X)
plt.scatter(X_mds[:,0], X_mds[:,1],c=labels)
```

```python
X_tsne = manifold.TSNE(n_components=2, init='random', learning_rate=100).fit_transform(X)
plt.scatter(X_tsne[:,0], X_tsne[:,1],c=labels)
```
